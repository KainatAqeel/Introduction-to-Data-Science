```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import csv
```

```python
data = pd.read_csv('world_pop.csv')
```

```python
data_2020 = data[['country', 'year_2020']].copy()
data_2020.set_index('country', inplace=True)
```

```python
data_2020.head(10)
```

```python
countries = data_2020.sort_values(by = "year_2020", ascending = False).head(10)
countries.head(10)
```

```python

plt.figure(figsize=(12, 8))
countries.plot(kind='bar', legend=None)
plt.title('Population of Top 10 Most Populated Countries in 2020', fontsize=18)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Population', fontsize=14)
plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=12)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
